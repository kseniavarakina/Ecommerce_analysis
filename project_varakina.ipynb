Дополнительно объевим функцию, чтобы удалить все ненужные символы

```python
def preprocessing_my(text):
    text = regex.sub('', text) # удаляем пунктуацию
    text = [token for token in text.split() if token not in russian_stopwords] # Удаляем стоп слова
    text = ''.join([token for token in text if not token.isdigit()]) 
    text = [token for token in text if token] # Удаляем пустые токены
    return ''.join(text)
```

##### Выделяем потребительские сегменты

Итак, мы поделили все товары на 30 тематических категорий, в которых слова сгруппированы по тематикам с помощью Latent Dirichlet allocation.

Формирую потребительские сегменты следующим образом:
    
    1. Ориентируемся на первое слово в каждой из выделенных тематик, т.к. оно имеет наибольший вес и практически везде радикально отличается от второго по значению слова.
    
    2. Первое слово каждой из тематик относим к более крупной категории товаров (растения / товары для дома / хранение и проч. - далее темы подробно перечислены). 

Таким образом, мы формируем потребительские сегменты по наиболее значимой категории товаров в тематике. 30 тематик распределить по смыслу легче, чем несколько тысяч товаров. Сегменты отличаются по тематике покупок - это значит, что мы сможем присвоить каждой покупке пользователя категорию, и на их основе определить поведение пользователей, которые покупают схожие категории товаров.

Итак - присваиваем каждой тематике более крупную категорию по смыслу, итоговое количество категорий сокращаем, исходя из схожести слов в текущих темах

#### Напишем функцию для присвоению каждому продукту категории

```python
def topic_prediction(my_document):
    string_input = [my_document]
    df = pd.DataFrame(string_input)
    X = vect.transform(df.stack())
    corpus = gensim.matutils.Sparse2Corpus(X, documents_columns=False)
    output = list(ldamodel_tfidf[corpus])[0]
    topics = sorted(output,key=lambda x:x[1],reverse=True)
    return topics[0][0]
 
```

Т.к. распределения не являются нормальными, для проверки статистических гипотез мы будем использовать критерий Манна-Уитни (непараметрический критерий, используемый для оценки различий между двумя независимыми выборками)

Каким образом будем проводить сравнения:
    
    1. Сначала используем критерий Манна-Уитни для вычисления pvalue
    2. Затем используем поправку Бонферрони
    3. На основе полученных значений pvalue проверим гипотезы

Сразу зададим функции поправки и проверки гипотез - будем использовать их для всех трех случаев.

```python
def bonferroni_correction(pvalues_list):
    p_adjusted = multipletests(pvalues_list, alpha=0.05, method='bonferroni')
    return(list(p_adjusted[1]))
```

```python
def hypothesis(pvalue):
    a=''
    if pvalue < alpha:
        a = 'Отклоняем нулевую гипотезу'
    else:
        a = 'Принимаем нулевую гипотезу'
    return(a)
```
